```python
import re

text = "I love playing football! It's amazing 😍 #AI"

# Step 1: lowercase
text = text.lower()

# Step 2: remove special characters (keep words)
text = re.sub(r"[^a-zA-Z0-9\s]", "", text)

# Step 3: split
tokens = text.split()

print(tokens)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uscased")

text = "I love playing football yash deliwala loves ai and ai coding"

tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print("Tokens:", tokens)
print("Token IDs:", token_ids)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

text = "AI is powerful how r u we yash deliwala go to bank.chat gpt is best.clause is best."

inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)
tokens = tokenizer.tokenize(text)
print(tokens)
print(len(tokens))

print(tokens[3])  # some word
print(outputs.last_hidden_state[0][3])

print(outputs.last_hidden_state.shape)


tokens_full = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])

for i, token in enumerate(tokens_full):
    print(i, token)
```

```python
import torch
import torch.nn as nn

# Example sentence: 5 words, each word = 10 features
x = torch.randn(1, 5, 10)  
# [batch, sequence_length, input_size]

lstm = nn.LSTM(input_size=10, hidden_size=20, batch_first=True)

output, (h_n, c_n) = lstm(x)

print("Output shape:", output.shape)
print("Hidden state shape:", h_n.shape)
```

```python
import torch
import torch.nn.functional as F

# pretend 3 words, each has 4 features
x = torch.randn(3, 4)

# similarity (who relates to whom)
scores = torch.matmul(x, x.T)

# convert to probabilities
weights = F.softmax(scores, dim=1)

# weighted sum
output = torch.matmul(weights, x)

print("Attention Weights:\n", weights)
print("Output:\n", output)
```
